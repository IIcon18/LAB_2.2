```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tample = pd.read_csv(r'C:\Users\ogkri\labIT\train.csv')
tample.dtypes.value_counts()
```

```python
tample.groupby(['Sex']).sum().plot(kind = 'bar', y = 'Survived')
plt.show()
```

```python
colours = ['green' if value == 0 else 'red' for value in tample['Survived']]

plt.scatter(tample["Age"], tample["Fare"], c=colours)
plt.gca().set(xlabel = 'Цена билета', ylabel = 'Возраст')
plt.title('Разница цены билета и возраста пассажира')
plt.show()
```
